```python
import numpy as np
def computeIDF(word, sentences):
    
    idf = {} # IDF mapping
    no_sentences = len(sentences)
    word_occurence_sentences = 0 # Number of sentences out of given sentences in 
                                 # which given word is present atleast once
                            
    
    for sentence in sentences:
        if(word in sentence):
            word_occurence_sentences += 1
    
    idf[word] = np.log(no_sentences/word_occurence_sentences)
    
    idf
        
```
