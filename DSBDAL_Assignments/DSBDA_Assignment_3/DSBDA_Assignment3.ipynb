```python
import pandas as pd
import numpy as np
df = pd.read_csv('nba.csv')
```

```python
df
```

Mean age of students studying in a college.

```python
result =df.groupby(['College']).mean()
mean_age_per_college = result["Age"]
mean_age_per_college
```

Median weight of team member in each team

```python
result =df.groupby(['Team']).median()
median_weight_per_team = result["Weight"]
median_weight_per_team
```

Minimum and Maximum salary in each college. 

```python
result =df.groupby(['College']).min()
min_salary_per_college = result["Salary"]
min_salary_per_college
```

```python
result =df.groupby(['College']).max()
max_salary_per_college = result["Salary"]
max_salary_per_college
```

```python
result =df.groupby(['Team']).std()
std_weight_per_team = result["Weight"]
```
