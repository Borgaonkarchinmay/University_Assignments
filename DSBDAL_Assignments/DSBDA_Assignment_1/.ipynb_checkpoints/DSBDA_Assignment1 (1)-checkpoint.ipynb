For sake of computational feasibility of our model, 
So we need to convert the categorical features of our data set
into numerical data.
In this case ('Species')

```python
df["species_code"] = pd.factorize(df.species)[0]
df
```

Displaying the species_code count.

```python
df.species_code.value_counts()
```

Implementing One Hot encoding

```python
df1 = pd.get_dummies(df["species"], drop_first = True)

```

```python
df = pd.concat([df, df1], axis = 1)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(16,10))
sns.heatmap(df.isna().transpose(),
            cmap="YlGnBu",
            cbar_kws={'label': 'Missing Data'})
plt.savefig("Missing_Data_Plot.png", dpi=100)
```

```python
perc = 70.0 # Like N %
min_count =  int(((100-perc)/100)*df.shape[0] + 1)

# Drop columns which have NaN values % >= 70%
df_new_dropped_cols = df.dropna( axis=1, thresh=min_count) 
df_new_dropped_cols.head()
```

```python
df_new_dropped_cols.isna().sum()
```

```python
cols = df_new_dropped_cols.columns
for col in cols:
  print(col)
```

```python
cols = df_new_dropped_cols.columns
col_list = []
for col in cols:
  col_list.append(col)

print(col_list)
```
