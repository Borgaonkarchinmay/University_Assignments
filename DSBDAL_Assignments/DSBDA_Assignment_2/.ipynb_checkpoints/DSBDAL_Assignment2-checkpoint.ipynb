Filling missing values

Part A) Fill with mean

```python
df.isna().sum()[df.isna().sum()>0].plot(kind='bar')
```

Before filling.
Columns having NaN values and their counts.

```python
df.isna().sum()
```

```python
# DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None, **kwargs)
df_mean_fill = df.fillna(df.mean)
```

```python
df_mean_fill.isna().sum()[df_mean_fill.isna().sum()>=0].plot(kind='bar')
```

After mean fillings

Part B) Replace with NaN's previous non-null value in the same column.

```python
df_forward_ffill = df.fillna(method = 'ffill')
```

```python
df_forward_ffill.isna().sum()[df_forward_ffill.isna().sum()>0].plot(kind='bar')
```

After forward filling

Part C) Replace with NaN's next non-null value in the same column.

```python
df_backward_bfill = df.fillna(method = 'bfill')
```

```python
if df_backward_bfill.isna().sum().empty:
  df_backward_bfill.isna().sum()[df_backward_bfill.isna().sum()>0].plot(kind='bar')
else: 
  print("No null values present")
```

Data standardization and Normalization

```python
sns.histplot(df)
```

Standardization. So that mean is 0 centered.

```python
standardized_df = (df - df.mean())/df.std() # Standard Scalar 
                                            # Z-score method
standardized_df
```

```python
sns.histplot(standardized_df)
```

Normalization on Health column.
This represents the health of students in
 the scale of 0 to 1.


```python
df["health"] = (df["health"] - df["health"].min())/(df["health"].max() - df["health"].min())
df
```

```python
sns.histplot(df["health"])
```

Outliers

Histogram for stu-por data set

```python
sns.histplot(df)
```

Data is positively skewed or right skewed.

```python
df["Dalc"].plot(kind='box')
```

```python
df["Dalc"].describe()
```

```python
Q1 = df["Dalc"].quantile(0.25)
Q3 = df["Dalc"].quantile(0.75)
Inter_Quantile_Range = Q3 - Q1
Inter_Quantile_Range
```

```python
Lower_Limit = Q1 - 1.5*Inter_Quantile_Range
Upper_Limit = Q3 + 1.5*Inter_Quantile_Range
Lower_Limit
```

```python
Upper_Limit
```

Handle outliers.

Outlier -> values beyond Inter-Quantile-Range

```python
below_lower_range = (df["Dalc"] < Lower_Limit)
above_upper_range = (df["Dalc"] < Upper_Limit)
df_outliers_removed = df[~(below_lower_range | above_upper_range)] # Dropping outliers
```

Histplot for Dalc : Before

```python
sns.histplot(df["Dalc"])
```

Histplot for Dalc : After

```python
sns.histplot(df_outliers_removed["Dalc"])
```

```python
df_outliers_removed["Dalc"].plot(kind = 'box')
```
