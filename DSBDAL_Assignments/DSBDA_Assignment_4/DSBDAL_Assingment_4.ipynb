```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('titanic.csv')
df
```

```python
sns.histplot(df["fare"])
```

```python
sns.scatterplot(data=df, x = "fare", y="pclass", hue="sex")
```

```python
sns.lineplot(data=df, x="pclass", y="alive")
```

```python
sns.barplot(data = df, x = "pclass", y="age", hue="alive")
```

```python
sns.swarmplot(data = df, x="pclass", y="fare", hue="alone")
```

```python
sns.histplot(data=df, x="pclass", hue="alive", multiple="stack")
```

```python
sns.boxplot(data=df, x="alive", y="fare", hue="sex")
```

```python
sns.violinplot(data=df, x="class", y="age", hue="alone", cut=0)
```

```python
sns.pointplot(data=df, x="class", y="age", hue="alive")
```
