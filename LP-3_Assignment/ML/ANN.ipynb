```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:, 3:12]
X
```

```python
Y = df.iloc[:, -1:]
Y
```

```python
df1 = pd.get_dummies(X["Gender"], drop_first = True)
df2 = pd.get_dummies(X["Geography"], drop_first = True)
df2
```

```python
X["Geography"].value_counts()
```

```python
X = pd.concat([X, df1, df2], axis = 1)
X
```

```python
X.drop(['Geography', 'Gender'], axis = 1, inplace = True)
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 40)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
classifier = Sequential()
```

```python
# Add first layer, which requires input size
classifier.add(Dense(6, activation = 'relu', input_shape = (X.shape[1], )))
classifier.add(Dropout(0.1))
```

```python
classifier.add(Dense(6, activation = 'relu'))
classifier.add(Dropout(0.1))
```

```python
# Add the output layer
classifier.add(Dense(1, activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = classifier.fit(x_train, y_train, batch_size = 32, epochs = 200, validation_split = 0.1, verbose = 2)
```

```python
classifier.summary()
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
y_pred = (y_pred > 0.5).astype(int)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```
